## NOTE: 
### In order to validate if the headlines were in fact 'fake' or 'real'

- headline: headline produced by an LLM
- exact: 'fake' or 'real' if the headline was a word for word match with a headline online
- story: 'fake' or 'real' if the headline was not a word for word match yet the general context of the headline could be associated with a news story.
- roberta: results from roberta in relation the the headlines
- bert-base-data: results from bert-base-uncased with the dataset it was originally fine-tuned with.
#### ^ BERT (1) - Bisaillon (2020) dataset. As written in thesis.
- bert-base-lstm: results from bert-base-uncased with the dataset fine-tuned with the dataset originally used on the LSTM model.
#### ^ BERT (2) - Lifferth (2018) dataset. As written in thesis.
- male-fake: number of male participants who tallied fake for the given headline.
- male-fake-percentage: percentage of male participants who tallied fake for the given headline.
- male-age: age of male participant.

The final 3 are the same respectively to female (male) and true (fake).

#### You will see an added column further down. This is called 'human-percent', by which I mean; the percentage of human participants who correctly identified the headlines as 'fake'.

## Analysing humans against ChatGPT3.5

```python
df_cgpt
```

```python
#filter and refine in accordance with the 'exact' column, which has all 'fake' strings in this case as a result of the headline confirmation.
#drop everything bar the ref, headline, 'exact' column and male/female 'fake' percentages
#https://saturncloud.io/blog/how-to-filter-pandas-dataframes-by-column-of-strings/#:~:text=To%20filter%20a%20DataFrame%20by,string%20value%20in%20the%20column.

filter_df_cgpt = df_cgpt[df_cgpt["exact"].str.contains("fake")]
filter_df_cgpt = df_cgpt.drop(['roberta', 'bert-base-data', 'bert-base-lstm', 'male-fake', 'female-fake', 'male-real', 'male-real-percent', 'female-real', 'female-real-percent', 'male-age', 'female-age', 'story'  ], axis=1)
filter_df_cgpt
```

```python
filter_df_cgpt['human-percent'] = filter_df_cgpt['male-fake-percent'] + filter_df_cgpt['female-fake-percent']
filter_df_cgpt
```

```python
#https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.pie.html
plot = filter_df_cgpt.plot.pie(y='human-percent', figsize=(5, 5))
```

```python
human_result_cgpt = filter_df_cgpt[["human-percent"]].mean()
print(human_result_cgpt)
```
